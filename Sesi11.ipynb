- Unsupervised (tidak punya target, contohnya untuk mengelompokkan data saja)
    - Clustering (mengelompokkan data jadi bbrp kelompok)
    - PCA Principal Component Analysis (mengurangi dimensi sebuah dataset)

# Regression
tipe machine learning modelling untuk data yang punya fitur dan punya target, target tipe angka. hasil model Machine Learning nya akan bisa membuat prediksi tipe angka.

## Linear Regression
linear - garis lurus

contohnya misalnya kita punya dataset luas rumah (fitur) dan harga rumah (target). apa hubungan antara luas rumah dan harga rumah. bisakah digambarkan dengan sebuah rumus seperti `harga rumah = luas rumah * 100 + 5000` ini adalah format formula matematik yang diberi nama linear regression. `y = ax + b`. tujuan linear regression adalah mencari formula ini, formula yang menggambarkan hubungan antara fitur dan target secara matematika.

## Polynomial Regression
polynomial - garis polynomial yang berbelok

seperti linear regression tapi formula matematika yang digunakan untuk mencari garis trend nya ada dalam bentuk polynomial `y = ax + bx2 + c` (contoh polynomial degree 2). kalau degree 3 ada kubik nya, formulanya ada `y = ax + bx2 + cx3 + d`, dan seterusnya.

```python
# import package
# untuk edit data
import numpy as np
import pandas as pd

# untuk visualisasi
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use(['ggplot'])

# untuk machine learning model nya
from sklearn.linear_model import LinearRegression
# untuk membuat polynomial regression kita akan pakai LinearRegression model dengan preprocessong Polynomial Features
from sklearn.preprocessing import PolynomialFeatures

# untuk mengambil contoh dataset dari sklearn
from sklearn import datasets

# untuk memisahkan data kita jadi 2 grup, training dan testing
from sklearn.model_selection import train_test_split
```

```python
# mengumpulkan data
# kita akan gunakan dummy data
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])

# visualisasi dummy data, anggap x luas rumah dan y harga rumah
plt.scatter(x, y)
plt.xlabel("luas rumah")
plt.ylabel("harga rumah")
plt.title("Dummy Dataset untuk Harga Rumah")
```

```python
# merapikan data
# karena data dummy, sudah rapi
# kita bisa langsung ke step berikut
```

```python
# membuat dan melatih model
model = LinearRegression()
model.fit(x, y)
```

```python
# evaluasi
model.score(x, y)
```

```python
# visualisasi (step optional)

# visualisasi dummy data, anggap x luas rumah dan y harga rumah
plt.scatter(x, y)
plt.xlabel("luas rumah")
plt.ylabel("harga rumah")

# visualisasi model kita
plt.scatter(x, model.predict(x), color="blue")
plt.plot(x, model.predict(x), color="blue")


plt.title("Linear Regression untuk Harga Rumah")
```

```python
# kita bisa pkai model untuk prediksi harga rumah dari luas rumah
# misal, luas rumah 50, berapa si harga rumahnya?
model.predict([[50]])
```

```python
# tuning
# untuk contoh yang ini belum ada step tambahan yang kita lakukan
```

# Polynomial Regression
mirip linear regression tapi formula matematiknya tipe polynomial

kita coba dengan data yang sama dulu, baru coba ke data tipe lain

```python
# mengumpulkan data
# kita akan gunakan dummy data
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])

# visualisasi dummy data, anggap x luas rumah dan y harga rumah
plt.scatter(x, y)
plt.xlabel("luas rumah")
plt.ylabel("harga rumah")
plt.title("Dummy Dataset untuk Harga Rumah")
```

```python
# data preparation
# ada step untuk merubah datanya jadi bentuk polynomial
transformer = PolynomialFeatures(degree=3, include_bias=False)
transformer.fit(x)

# buat data kita yang versi polynomial
x_poly = transformer.transform(x)

print(x_poly)
```

```python
# buat dan latih model
model = LinearRegression().fit(x_poly, y)
```

```python
# evaluasi model
model.score(x_poly, y)
```

```python
# tuning
# untuk polynomial regression, kita bisa memilih mau polynomial regression degree berapa dengan mudah
```

```python
# visualisasi

# visualisasi dummy data, anggap x luas rumah dan y harga rumah
plt.scatter(x, y)
plt.xlabel("luas rumah")
plt.ylabel("harga rumah")

# visualisasi model kita
dummy_x = np.array(range(0,60)).reshape((-1,1))
dummy_x_poly = transformer.transform(dummy_x)

plt.scatter(x, model.predict(x_poly), color="blue")
plt.plot(dummy_x, model.predict(dummy_x_poly), color="blue")

plt.title("Polynomial Regression untuk Harga Rumah")
```

## Contoh 2: Polynomial Regression

# Linear Regression dengan Real Dataset

```python
# siapin data
data_boston = datasets.load_boston()
# data_boston.data # datanya sendiri
# data_boston.feature_names # judul kolom
# print(data_boston.DESCR) # deskripsi datanya
# data_boston.target # targetnya untuk dataset ini adalah harga rumah

# kugabung jadi sebuah pandas dataframe
df_boston = pd.DataFrame(data = data_boston.data, columns=data_boston.feature_names)
df_boston['price'] = data_boston.target

df_boston
```

### Linear Regression

```python
# rapikan data
fitur = df_boston[data_boston.feature_names]
target = df_boston['price']
```

```python
# bikin dan latih model
model = LinearRegression().fit(fitur, target)
```

```python
# evaluasi
model.score(fitur, target)
```

### Polynomial Regression

```python
# rapikan data
fitur = df_boston[data_boston.feature_names]
target = df_boston['price']

transformer = PolynomialFeatures(degree=2, include_bias=False)
transformer.fit(fitur)

fitur_poly = transformer.transform(fitur)

# bikin dan latih model
model = LinearRegression().fit(fitur_poly, target)

# evaluasi
model.score(fitur_poly, target)
```

## Step tambahan di merapikan data
split data jadi training dan testing data

```python
# Linear Regression, pakai split data
# rapikan data
fitur = df_boston[data_boston.feature_names]
target = df_boston['price']

# split jadi training dan testing data
x_train, x_test, y_train, y_test = train_test_split(fitur, target, random_state=0)
# opsi random_state adalah untuk memastikan kita selalu mengambil data training dan testing yang sama
print(x_train.shape, x_test.shape)

# bikin dan latih model
model = LinearRegression().fit(x_train, y_train)

# evaluasi
model.score(x_test, y_test)
```

```python
# Polynomial Regression, pakai split data
# rapikan data
fitur = df_boston[data_boston.feature_names]
target = df_boston['price']

transformer = PolynomialFeatures(degree=2, include_bias=False)
transformer.fit(fitur)

fitur_poly = transformer.transform(fitur)

# split data jadi training dan testing data
x_train, x_test, y_train, y_test = train_test_split(fitur, target, random_state=0)
x_train_poly = transformer.transform(x_train)
x_test_poly = transformer.transform(x_test)

# bikin dan latih model
model = LinearRegression().fit(x_train_poly, y_train)

# evaluasi
model.score(x_test_poly, y_test)
```
