```python
# import package
from wordcloud import WordCloud, STOPWORDS
import numpy as np
import pandas as pd

# untuk import image menjadi array
from PIL import Image

# untuk visualisasi
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# import data
# baca file txt jadi string pakai basic python fungsi open
alice_novel = open('dataset/alice_novel.txt', 'r').read()

print(len(alice_novel))
```

```python
# buat wordcloud
wc_alice = WordCloud()
wc_alice.generate(alice_novel)
```

```python
# tampilkan wordcloud secara visual
plt.imshow(wc_alice)
```

```python
# tanpa axis
plt.imshow(wc_alice)
plt.axis("off")
```

```python
# mau buat wordcloud yang backgroun putih
# buat wordcloud
wc_alice = WordCloud(background_color="white")
wc_alice.generate(alice_novel)

plt.imshow(wc_alice)
plt.axis("off")
```

```python
# kita gunakan opsi stopwords untuk exclude kata kata yang tidak mau ditampilkan di word cloud
sw = ["said"]

wc_alice = WordCloud(background_color="white",
                    stopwords = sw)
wc_alice.generate(alice_novel)

plt.imshow(wc_alice)
plt.axis("off")
```

```python
# kita gunakan list stopwords dari package worldcloud juga
print(STOPWORDS, type(STOPWORDS), len(STOPWORDS))
```

```python
# stopwords yang sudah kita import, kita ubah jadi list, lalu kita tambahkan "said" tadi
sw = list(STOPWORDS)
sw.append("said")
print(sw, type(sw), len(sw))
```

```python
# buat lagi word cloud nya dengan stopwords lebih banyak (193 kata)
wc_alice = WordCloud(background_color="white",
                    stopwords = sw)
wc_alice.generate(alice_novel)

plt.imshow(wc_alice)
plt.axis("off")
```

```python
# kali ini kita ubah width dan height word cloud nya
wc_alice = WordCloud(background_color="white",
                    stopwords = sw,
                    width=2000,
                    height=1000)
wc_alice.generate(alice_novel)

# pakai figure
fig = plt.figure()
fig.set_figwidth(18)
fig.set_figheight(9)

# tambahkan interpolation supaya edges gambarnya lebih halus
# detail interpolation ada opsi apa saja dan efeknya bisa dilihat di 
# https://matplotlib.org/stable/gallery/images_contours_and_fields/interpolation_methods.html
plt.imshow(wc_alice, interpolation="bilinear")
plt.axis("off")
```

# Word Cloud Pakai Mask
bentuk word cloud tidak harus kotak/ persegi, bisa bentuk apapun. siapkan dulu gambar hitam diatas putih (umumnya gambar mask) sebagai base bentuk word cloud nya.

```python
# download dulu gambar mask alice nya
!wget --quiet https://github.com/ardhiraka/PFDS_sources/raw/master/alice_mask.png -O alice_mask.png
```

```python
# import gambar dan ubah jadi numpy array
mask_alice = np.array(Image.open("alice_mask.png"))
```

```python
wc_alice = WordCloud(background_color="white",
                    stopwords = sw,
                    width=2000,
                    height=1000,
                    mask=mask_alice)
wc_alice.generate(alice_novel)

fig = plt.figure()
fig.set_figwidth(18)
fig.set_figheight(9)

plt.imshow(wc_alice, interpolation="bilinear")
plt.axis("off")
```

# Word Cloud dari data table
misalkan data imigrasi Canada yang kita gunakan kemarin di sesi 7 (bukan novel, cerita, atau lainnya)

kita bisa bikin wordcloud negara mana yang paling banyak mengirim imigran ke Canada secara keseluruhan 1980-2013 font nya paling besar di wordcloud, yang jarang2 maskin kecil atau bahkan tidak muncul di wordcloud

kita akan onvert data table misalnya China 100, Indonesia 2 --> text "China China China... Indonesia Indonesia" baru kita jadikan wordcloud

```python
# import data
df = pd.read_excel("dataset/Canada.xlsx", 
                   sheet_name="Canada by Citizenship",
                   skiprows=20,
                   skipfooter=2)

# rapiiin data
hapus_kolom = ['Type', 'Coverage', 'AREA', 'REG', 'DEV', 'DevName']
df.drop(columns=hapus_kolom, inplace=True)

df.rename(columns={"OdName":"Country", "AreaName":"Continent", "RegName":"Region"}, inplace=True)

years = range(1980, 2014)

df.set_index("Country", inplace=True)

df.head()
```

```python
# buat dulu kolom total, karena kita mau buat wordcloud berdasarkan total imigran dari setiap negara selama 1980-2013
df['Total']=df[years].sum(axis=1)
df.head()
```

```python
# drop lagi kolom yang tidak dipakai
df.drop(years, axis=1, inplace=True)
df.head()
```

```python
print(df.shape)
```

```python
# kita mau convert total imigran menjadi berapa kali nama negara tersbut muncul di kumpulan text kita
# tapi kalau tidak dibuat per rasio, bisa panjang sekali text kita
total_immigration = df['Total'].sum()
total_immigration
```

```python
# supaya lebih efisien, kita bagi dulu per ratio total imigran, setiap negara termasuk berapa kali namanya muncul
# kita kasih limit mau berapa kata yang muncul
max_words = 100

kumpulan_can = ""

# kita hitung dengan rasio, setiap negara jadi muncul berapa kali namanya
for country in df.index.values:
    nama_country = country.replace(" ", "")
    nama_country = nama_country.replace(",","")
    nama_country = nama_country.replace("(","")
    nama_country = nama_country.replace(")","")
    nama_country = nama_country.replace("'","")
    
    repeat = int(df.loc[country, 'Total']/float(total_immigration)*max_words)
    # print(nama_country, repeat)
    
    kumpulan_can = kumpulan_can + (nama_country + " ") * repeat
    
print(kumpulan_can)
```
